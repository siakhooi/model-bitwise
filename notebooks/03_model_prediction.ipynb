# Model Prediction

Load the XOR model and make predictions using inference input data.

```python
import pandas as pd
import tensorflow as tf
```

## Load the input data

```python
input_df = pd.read_csv('../inference/input/input.csv')
print("Input data:")
input_df
```

## Load the Keras model

```python
model = tf.keras.models.load_model('../models/xor_model.keras')
model.summary()
```

## Prepare input data for prediction

```python
X = input_df[['X1', 'X2']].values
```

## Predictions

```python
predictions = model.predict(X)
```

## Process Results

```python

# Add predictions to the dataframe
input_df['Prediction'] = predictions

# Convert probabilities to class labels (0 or 1)
input_df['Predicted_Class'] = (predictions > 0.5).astype(int)

```

## Predictions

```python
input_df
```

## Save Predictions

### Save to CSV File

```python
output_csv_path = '../inference/output/output.csv'
input_df.to_csv(output_csv_path, index=False)
```

### Save to JSON File

```python
output_json_path = '../inference/output/output.json'

input_df.to_json(output_json_path, orient='records', indent=2)
```

### Summary

```python
print(f"Predictions saved to:\n  - {output_csv_path}\n  - {output_json_path}")
```
